```python
import pandas as pd 
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])


#get jobs in the US
df_US = df[df['job_country'] == 'United States']
df_US_jobs = df_US.groupby('job_title_short')['salary_year_avg'].mean().sort_values(ascending=False)
df_US_jobs
```

```python
df_count = df_US['job_title_short'].value_counts()
df_count
```

```python
df_salary = df_US.groupby('job_title_short')['salary_year_avg'].agg(['median', 'min', 'max', 'count']).sort_values(by='median', ascending=False)
df_salary
```
